```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
df = pd.read_csv("/content/placement (5).csv")
print(df)
```

```python
df.head()
```

```python
# Columns aur data types dekhna
print(df.info())

# Missing values check karna
print(df.isnull().sum())

# Summary statistics dekhna (Mean, Std, Min, Max, etc.)
print(df.describe())
```

```python
df.shape
```

```python
df = df.drop(columns=["Unnamed: 0"])

```

```python
# Input Features (X) aur Target Variable (y)
X = df[["cgpa", "iq"]]  # Input (Features)
y = df["placement"]  # Output (Target)

# Shapes check karna
print("X Shape:", X.shape)
print("y Shape:", y.shape)

# Pehle 5 rows dekhna
print(X.head())
print(y.head())
```

```python
y.shape
```

```python
plt.scatter(df["cgpa"], df["iq"], c=df["placement"], cmap="coolwarm", alpha=0.7)

# Labels aur Title
plt.xlabel("CGPA")
plt.ylabel("IQ")
plt.title("Scatter Plot of CGPA vs IQ (Colored by Placement)")


# Show Plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
print("X_train Shape:", X_train.shape)
print("y_train Shape:", y_train.shape)
print("X_test Shape:", X_test.shape)
print("y_test Shape:", y_test.shape)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train_scaled, y_train)
```

```python
y_pred = model.predict(X_test_scaled)
```

```python

from sklearn.metrics import accuracy_score, confusion_matrix
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
```

```python
import joblib

# Model ko "model.pkl" file mein save karna
joblib.dump(model, "model.pkl")

print("✅ Model successfully saved as model.pkl")  # ✅ String properly closed
```
